```python
from langgraph.graph import StateGraph,  START , END
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatPerplexity()
```

```python
class llmstate(TypedDict):
    question:str
    answer:str
```

```python
def questionnode(state:llmstate)->llmstate:
    question= state['question']

    prompt = f'Anser the question properly and make user happy {question}'
    answer = model.invoke(prompt).content

    state['answer'] = answer

    return state
```

```python
# Create a graph

graph = StateGraph(llmstate)

graph.add_node('questionnode',questionnode)

graph.add_edge(START,'questionnode')
graph.add_edge('questionnode',END)

workflow = graph.compile()
```

```python
initial_state = {
    'question': 'How is the weather today at TAMLUK, WB ?',
 "answer": ""
''}

final_state = workflow.invoke(initial_state)

print(final_state['answer'])


```
