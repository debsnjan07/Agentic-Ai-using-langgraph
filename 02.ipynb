```python
from langgraph.graph import StateGraph, START ,END

from typing import TypedDict,Optional
```

```python
class Condition(TypedDict):
    inputone = int
    inputtwo =int
    operation = str
    output = Optional
```

```python
def addition(state:Condition) -> Condition:
    one = state['inputone']
    two = state['inputtwo']

    operation = one+two 
    state['output'] = operation 

    return state

def subtraction(state:Condition) ->Condition:
    one = state['inputone']
    two = state['inputtwo']

    operation = one-two 
    state['output'] = operation 
    return state

def Router(state:Condition)->str:
    if state['operation'] =="+":
        return addition
    elif state['operation'] =="-":
        return subtraction
    else :
     return False
    
```

```python
graph = StateGraph(Condition)

graph.add_node("addition",addition)
graph.add_node("subtraction",subtraction)
graph.add_node("router",Router)

graph.add_edge(START,'router')

graph.add_conditional_edges(
    'router',
    Router,
    {
        "addition_operation":'addition',
        "subtract_operation":'subtraction',
    }
)
graph.add_edge('addition',END)
graph.add_edge('subtraction',END)

app = graph.compile()

```

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Optional

class Condition(TypedDict):
    inputone: int
    inputtwo: int
    operation: str
    output: Optional[int]

def addition(state: Condition) -> Condition:
    state["output"] = state["inputone"] + state["inputtwo"]
    return state

def subtraction(state: Condition) -> Condition:
    state["output"] = state["inputone"] - state["inputtwo"]
    return state


def route_operation(state: Condition) -> str:
    if state["operation"] == "+":
        return "addition_operation"
    elif state["operation"] == "-":
        return "subtract_operation"
    else:
        raise ValueError("Unsupported operation")

graph = StateGraph(Condition)

graph.add_node("addition", addition)
graph.add_node("subtraction", subtraction)

graph.add_conditional_edges(
    START,          
    route_operation,
    {
        "addition_operation": "addition",
        "subtract_operation": "subtraction",
    },
)

graph.add_edge("addition", END)
graph.add_edge("subtraction", END)

app = graph.compile()

initial_state: Condition = {
    "inputone": 56,
    "inputtwo": 90,
    "operation": "+",
    "output": None,
}

output = app.invoke(initial_state)
print(output)
```
